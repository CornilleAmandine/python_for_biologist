```python
#Q1 - Find the 10 largest chromosomes with respect to the number of base pairs
select1= hg[["Chromosome", "Base_Pairs"]]
select1
select2=select1.sort_values(by=["Base_Pairs"], ascending=False)
select2.head(10) #Take the first 10 hiet
```

```python
#Q2 - enlist chromosomes with >=1000 protein coloding genes and > 500 pseudogenes
#Chr_genes2 = hg[["Chromosome", "Protein_Coding_genes", "Pseudogenes"]]
hg.loc[(hg["Protein_Coding_genes"] >=1000) & (hg["Pseudogenes"]>=500)]
```

```python
#Q3: List the first 10 chromosomes with the highest density
hg2=hg
hg2["Gene_density_per_megabase"] =  (hg["Protein_Coding_genes"]/hg["Base_Pairs"]) * 1000000
hg2
hg3=hg2.sort_values(by=["Gene_density_per_megabase"], ascending=False)
#hg3.head(10) #Take the first 10 hiet
hg3.iloc[0:10] #Take the first 10 hiet
```

```python
#Q4 - Calculate the mena gene dentisty for autsome, sex chromsosomes and mitochondria
autosomes=hg2.iloc[0:22] #22 because it is exclusive
autosomes[["Gene_density_per_megabase"]].mean()
```

```python
sexchromosomes=hg2.iloc[22:24]
sexchromosomes[["Gene_density_per_megabase"]].mean()
```
